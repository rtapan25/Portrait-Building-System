```python
import glob
import pandas as pd
import json
import numpy as np
```

```python
features= {'eyes':5, 'eye_brows':5, 'lips': 5, 'nose':5}
```

```python
df = pd.DataFrame (columns = ['Image','eyes','eye_brows','lips','nose', 'face_shape'])
```

```python
clusters={}
for i in features:
    clusters[i]=[0]*features[i]
    for j in range(features[i]):
        clusters[i][j]=[]
        for file in glob.glob(i+'/output/cluster'+str(j)+'/*.jpg'):
            clusters[i][j].append(file[file.rfind('\\')+1:])
```

```python
clusters['eyes'][4]
```

```python
df['Image']=range(9601)
```

```python
for i in df['Image']:
    for feat in clusters:
        for j in range(len(clusters[feat])):
            if str(i)+'.jpg' in clusters[feat][j]:
                df[feat][i]=j
                break
    
```

```python
with open('C:/Users/tapan/Downloads/attributes_face_shape.txt', 'r') as reader:
    f = [line for line in reader.readlines()]
for i in range(len(f)):
    f[i] = f[i].replace("'", '"')
    f[i] = f[i].replace(".jpg", '')
```

```python
kao=[]
for i in f:
    kao.append(json.loads(i))
face={}
for i in kao:
    face[int(i['image'])]=i['face_shape']
    
```

```python
for i in df['Image']:
    df['face_shape'][i] = face[i]
```

```python
df.to_csv('im_clust5.csv')
```
